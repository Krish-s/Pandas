```python
#defining independent and dependent variables
x1 =df["Exam"] # - independent value always in machine learning
y = df["GPA"]  # - depedent value always in machine learning

#y^ = predicted dependent value
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.scatter(x1,y)
plt.show()
```

```python
import statsmodels.api as sm
x= sm.add_constant(x1)
x
```

```python
model = sm.OLS(y,x)
```

```python
result = model.fit()
result.summary()
```
