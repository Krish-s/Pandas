```python
df["total_rooms_new1"] = stats.boxcox(df["total_rooms"],lmbda = 0)
df
```

```python
#lmbda = 0 = log(x)

#lmbda = 0.5 = sqrt(x)

#lmbda = -0.5 = 1/sqrt(x)

#lmbda = 1 = x

#lmbda = -1 = 1/x

#lmbda = 2 = x**2

#lmbda = -2 = 1/x**2
```

```python
for i in df.columns:
  if df[i].skew() < 0.5 and df[i].skew() > -0.5 and df[i].kurtosis() < 3 and df[i].kurtosis() > -3 :
    continue
  print(i)
```

```python
for i in df.columns:
  if df[i].skew() < 0.5 and df[i].skew() > -0.5 and df[i].kurtosis() < 3 and df[i].kurtosis() > -3 :
    continue
  print(i)
  method = [0,0.5,1,-1,2,-2]
  for j in method:
    print(j)
    print("skewness",pd.DataFrame(stats.boxcox(df[i],lmbda =j)).skew().values,"Kurtosis",pd.DataFrame(stats.boxcox(df[i],lmbda = j)).kurtosis().values)
  print("-------------------")
```
