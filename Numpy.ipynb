```python
tips = sns.load_dataset("tips")
tips
```

```python
data = pd.crosstab(tips["smoker"],tips["sex"])
data
```

```python
data = pd.crosstab(tips['size'],tips["day"])
data
```

```python
data.values
```

```python
observed_value = data.values
observed_value
```

```python
value = stats.chi2_contingency(data)
value
```

```python
expected_value = value[3]
expected_value
```

```python
#contingency understanding
sum([((o-e)**2)/e for o,e in zip(observed_value,expected_value)])
```

```python
our_chi2=sum(sum([((o-e)**2)/e for o,e in zip(observed_value,expected_value)]))
our_chi2
```

```python
#level of significance 2
critical_point = stats.chi2.ppf(q=0.98,df=15)
critical_point
```

```python
critical_point = stats.chi2.ppf(q=0.98,df=15)
if our_chi2 > critical_point:
  print(critical_point)
  print("H0 null hypothesis rejected,Ha alternate hypothesis accepted,data is dependent")
else:
  print(critical_point)
  print("H0 null hypothesis accepted,Ha alternate hypothesis rejected,data is Independent")
```

```python
p_value = stats.chi2.cdf(x=our_chi2,df=15)
p_value
```

```python
p_value=1-stats.chi2.cdf(x=our_chi2,df=15)
print(p_value)
if p_value > 0.02:
  print("H0 null hypothesis accepted,Ha alternate hypothesis rejected,data is Independent")
else:
  print("H0 null hypothesis rejected,Ha alternate hypothesis accepted,data is dependent")
```
