```python
expected_value = value[3]
expected_value
```

```python
#contingency understanding
sum([((o-e)**2)/e for o,e in zip(observed_value,expected_value)])
```

```python
our_chi2=sum(sum([((o-e)**2)/e for o,e in zip(observed_value,expected_value)]))
our_chi2
```

```python
#level of significance 2
critical_point = stats.chi2.ppf(q=0.98,df=15)
critical_point
```

```python
critical_point = stats.chi2.ppf(q=0.98,df=15)
if our_chi2 > critical_point:
  print(critical_point)
  print("H0 null hypothesis rejected,Ha alternate hypothesis accepted,data is dependent")
else:
  print(critical_point)
  print("H0 null hypothesis accepted,Ha alternate hypothesis rejected,data is Independent")
```

```python
p_value = stats.chi2.cdf(x=our_chi2,df=15)
p_value
```

```python
p_value=1-stats.chi2.cdf(x=our_chi2,df=15)
print(p_value)
if p_value > 0.02:
  print("H0 null hypothesis accepted,Ha alternate hypothesis rejected,data is Independent")
else:
  print("H0 null hypothesis rejected,Ha alternate hypothesis accepted,data is dependent")
```

```python
#two sample test

def two_sample(d1,d2):
  t=0
  f=0
  for i in range(31):
    sample1 = d1.sample(frac = 0.03)
    sample2 = d2.sample(frac =0.03)
    t_test,p_value= stats.ttest_ind(sample1,sample2)
    if p_value < 0.55:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False
```

```python
two_sample(df['INT_SQFT'],df['DIST_MAINROAD'])
```

```python
#chi2 contigency for category vs category

def chi2con(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.55 else False
```

```python
chi2con(df["AREA"],df["N_BEDROOM"])
```

```python
#oneway annova for continuous vs category

def annova_test(d1,d2):
  group = df[d2].unique()
  data = {}
  for i in group:
    data[i]=df[d1][df[d2]==i]
  f_test,f_value = stats.f_oneway(*[i for i in data.values()])
  if p_value < 0.55:
    return  False
  else:
    return True
```

```python
annova_test('INT_SQFT','AREA')
```

```python
final={}
for i in df.columns:
  final[i]={}
  for j in df.columns:
    if (i in continues) and (j in continues):
      result = two_sample(df[i],df[j])
    elif (i in continues) and (j in categories):
      result = annova_test(i,j)
    elif(i in categories) and (j in continues):
      result = annova_test(j,i)
    elif(i in categories) and (j in categories):
      result = chi2con(df[i],df[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
```
