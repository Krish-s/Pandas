```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
a = np.random.randint(1,100,150)
b = np.random.randint(1,100,150)
```

```python
a
```

```python
b
```

```python
import plotly.express as px
```

```python
px.area(x=a,y=b)
```

```python
px.bar(x=a,y=b)
```

```python
px.bar_polar(a,b)
```

```python
px.box(a,b)
```

```python
px.scatter(x=a,y=b)
```

```python
px.scatter(x=a,y=b,labels= {'x':'a value','y':'b value'},title='a,b value scatter plot')
```

```python
px.line(x=a,y=b,labels={'x':'a value','y':'b value'},title = 'a,b values')
```

```python
#in order data

c = np.arange(1,151)
px.line(x=c,y=b)

# all arguements should have same data length
```

```python
px.area(x=c,y=b,title ="area ordered cont, random int")
```

```python
px.bar(x=c,y=b,title ="bar")
```

```python
px.scatter(x=c,y=b,title ="scatter ordered cont, random int")
```
