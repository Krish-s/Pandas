```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.express as px
```

```python
df = px.data.tips()
df
```

```python
px.histogram(df,x="total_bill",nbins = 20, color ="sex")
```

```python
px.pie(df,names="day",values = "total_bill",hole = 0.4)
```

```python
df2 = px.data.gapminder().query("year == 2007")
df2
```

```python
px.histogram(df2,x="gdpPercap",nbins = 20,color="continent")
```

```python
px.pie(df2,names= "continent",values="gdpPercap")
```

```python
fig = px.pie(df2,names="continent",values="gdpPercap")
fig.update_traces(textposition="outside")
fig.show()
```

```python
px.sunburst(df,path =["sex","smoker","day","size"],names="total_bill ")
```
